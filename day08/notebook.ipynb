## ðŸ“š Topic Introduction

**Conditional statements** allow your program to make decisions and execute different code based on conditions.

Python conditional keywords:
- **if**: Execute code if condition is True
- **elif**: Check another condition if previous was False
- **else**: Execute code if all conditions are False

---

## ðŸ” Syntax & Rules

### Basic If Statement:
```python
if condition:
    # code to execute
```

### If-Else:
```python
if condition:
    # code if True
else:
    # code if False
```

### If-Elif-Else:
```python
if condition1:
    # code
elif condition2:
    # code
else:
    # code
```

**Important**: Indentation matters in Python!

---

## ðŸ’¡ Examples

````python
# Example 1: Basic if statement
age = 20

if age >= 18:
    print("You are an adult")
    print("You can vote")
````

````python
# Example 2: If-else statement
temperature = 25

if temperature > 30:
    print("It's hot outside!")
else:
    print("The weather is pleasant")
````

````python
# Example 3: If-elif-else
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your grade is: {grade}")
````

````python
# Example 4: Nested conditionals
age = 25
has_license = True

if age >= 18:
    if has_license:
        print("You can drive")
    else:
        print("You need a license")
else:
    print("You are too young to drive")
````

````python
# Example 5: Multiple conditions with and/or
age = 25
has_license = True
has_car = True

if age >= 18 and has_license and has_car:
    print("You can drive your car")
elif age >= 18 and has_license:
    print("You can drive but need a car")
else:
    print("You cannot drive")
````

````python
# Example 6: Checking if value is in a list
fruits = ["apple", "banana", "cherry"]
search = "banana"

if search in fruits:
    print(f"{search} is available")
else:
    print(f"{search} is not available")
````

````python
# Example 7: Ternary operator (one-line if-else)
age = 20
status = "adult" if age >= 18 else "minor"
print(f"Status: {status}")
````

````python
# Example 8: Checking multiple values
day = "Saturday"

if day in ["Saturday", "Sunday"]:
    print("It's the weekend!")
else:
    print("It's a weekday")
````

## ðŸŽ¯ Challenge Problems

````python
# Challenge 1: Triangle validator
# Check if three sides can form a valid triangle
# Rule: Sum of any two sides must be greater than the third

# Your code here
side1 = 3
side2 = 4
side3 = 5
if (side1 + side2 > side3) and (side1 + side3 > side2) and (side2 + side3 > side1):
    print("The sides can form a valid triangle")
else:
    print("The sides cannot form a valid triangle")
````

````python
# Challenge 2: Quadratic equation solver
# Given a, b, c, determine if equation has real solutions
# Check discriminant (bÂ²-4ac): positive (2 solutions), zero (1), negative (none)

# Your code here
import math
a = 1
b = -3
c = 2
discriminant = b**2 - 4*a*c
if discriminant > 0:
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"The equation has two real solutions: {root1} and {root2}")
elif discriminant == 0:
    root = -b / (2*a)
    print(f"The equation has one real solution: {root}")
else:
    print("The equation has no real solutions")
````

````python
# Challenge 3: Rock, Paper, Scissors game
# Two players choose, determine the winner

# Your code here
player1 = "rock"
player2 = "scissors"
if player1 == player2:
    print("It's a tie!")
elif (player1 == "rock" and player2 == "scissors") or \
     (player1 == "scissors" and player2 == "paper") or \
     (player1 == "paper" and player2 == "rock"):
    print("Player 1 wins!")
else:
    print("Player 2 wins!")
````

````python
# Challenge 4: Tax calculator with brackets
# 0-10000: 10%, 10001-30000: 20%, 30001+: 30%

# Your code here
income = 45000
if income <= 10000:
    tax = income * 0.10
elif income <= 30000:
    tax = 10000 * 0.10 + (income - 10000) * 0.20
else:
    tax = 10000 * 0.10 + 20000 * 0.20 + (income - 30000) * 0.30
print(f"The total tax is: ${tax:.2f}")
````

````python
# Challenge 5: Date validator
# Check if day, month, year form a valid date
# Consider leap years and days in each month

# Your code here
day, month, year = 29, 2, 2024
valid = False

if month in [1, 3, 5, 7, 8, 10, 12]:
    valid = 1 <= day <= 31
elif month in [4, 6, 9, 11]:
    valid = 1 <= day <= 30
elif month == 2:
    is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
    valid = 1 <= day <= (29 if is_leap else 28)

if valid:
    print("Valid date")
else:
    print("Invalid date")
````
