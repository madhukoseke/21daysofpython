Welcome to Day 9! Today we'll learn about for loops - a powerful way to repeat code.

---

## üìö Topic Introduction

**For loops** allow you to iterate over a sequence (list, tuple, string, range) and execute code for each item.

Common uses:
- Iterate through lists, tuples, strings
- Repeat code a specific number of times
- Process each element in a collection

---

## üîç Syntax & Rules

### Basic For Loop:
```python
for item in sequence:
    # code to execute
```

### Using range():
```python
range(stop)           # 0 to stop-1
range(start, stop)    # start to stop-1
range(start, stop, step)  # with step
```

---

## üí° Examples

````python
# Example 1: Iterating through a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")
````

````python
# Example 2: Using range()
print("Count to 5:")
for i in range(1, 6):
    print(i)
````

````python
# Example 3: Iterating through a string
word = "Python"
for letter in word:
    print(letter)
````

````python
# Example 4: Using enumerate()
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
````

````python
# Example 5: Nested for loops
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i={i}, j={j}")
````

````python
# Example 6: Iterating through a dictionary
person = {"name": "Alice", "age": 25, "city": "Boston"}
for key, value in person.items():
    print(f"{key}: {value}")
````

````python
# Example 7: Using range() with step
print("Even numbers from 0 to 10:")
for i in range(0, 11, 2):
    print(i)
````

````python
# Example 8: List comprehension (compact for loop)
numbers = [1, 2, 3, 4, 5]
squared = [x**2 for x in numbers]
print(squared)
````
