### Challenge Problem Solutions

```python
# Challenge 1 Solution
students = {"Alice": 92, "Bob": 78, "Charlie": 85, "David": 95, "Eve": 88}

# Average
average = sum(students.values()) / len(students)
print(f"Average score: {average:.2f}")

# Highest scorer
highest = max(students, key=students.get)
print(f"Highest scorer: {highest} with {students[highest]}")

# Students above 80
above_80 = [name for name, score in students.items() if score > 80]
print(f"Students above 80: {above_80}")
```

```python
# Challenge 2 Solution
sentence = "python is great and python is fun"
words = sentence.lower().split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
print(frequency)
```

```python
# Challenge 3 Solution
numbers = list(range(1, 11))

# Even numbers
evens = [n for n in numbers if n % 2 == 0]
print(f"Even numbers: {evens}")

# Sum of odd numbers
odd_sum = sum([n for n in numbers if n % 2 != 0])
print(f"Sum of odd numbers: {odd_sum}")

# Divisible by 3
div_by_3 = [n for n in numbers if n % 3 == 0]
print(f"Divisible by 3: {div_by_3}")
```

```python
# Challenge 4 Solution
contacts = {
    "Alice": {"phone": "555-0001", "email": "alice@email.com"},
    "Bob": {"phone": "555-0002", "email": "bob@email.com"},
    "Charlie": {"phone": "555-0003", "email": "charlie@email.com"}
}

# Search
search_name = "Bob"
if search_name in contacts:
    print(f"{search_name}: {contacts[search_name]}")
```

```python
# Challenge 5 Solution
mixed = [1, "hello", 3.14, "world", 42, 2.718, "python", 100]
integers = [item for item in mixed if type(item) == int]
strings = [item for item in mixed if type(item) == str]
floats = [item for item in mixed if type(item) == float]

print(f"Integers: {integers}")
print(f"Strings: {strings}")
print(f"Floats: {floats}")
```
