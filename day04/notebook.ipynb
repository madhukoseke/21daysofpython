## ðŸ“š Topic Introduction

### Tuples
**Tuples** are ordered, immutable collections. Once created, they cannot be modified.

Key characteristics:
- **Ordered**: Maintain their order
- **Immutable**: Cannot be changed after creation
- **Allow duplicates**: Can contain duplicate values
- **Faster than lists**: More efficient for read-only data

### Sets
**Sets** are unordered collections of unique elements.

Key characteristics:
- **Unordered**: No specific order
- **Unique elements**: Automatically removes duplicates
- **Mutable**: Can add/remove elements
- **Fast lookups**: Very efficient for membership testing

---

## ðŸ” Syntax & Rules

### Tuples:
```python
empty_tuple = ()
single_item = (1,)  # Note the comma
my_tuple = (1, 2, 3)
```

### Sets:
```python
empty_set = set()  # NOT {}
my_set = {1, 2, 3}
```

## ðŸ’¡ Examples - Tuples

````python
# Example 1: Creating tuples
empty_tuple = ()
single_item = (42,)  # Comma is required
coordinates = (10, 20)
rgb_color = (255, 128, 0)
mixed_tuple = (1, "hello", 3.14, True)

print("Empty:", empty_tuple)
print("Single:", single_item)
print("Coordinates:", coordinates)
print("RGB:", rgb_color)
print("Mixed:", mixed_tuple)
````

````python
# Example 2: Accessing tuple elements
fruits = ("apple", "banana", "cherry", "date")

print(f"First: {fruits[0]}")
print(f"Last: {fruits[-1]}")
print(f"Slice: {fruits[1:3]}")
````

````python
# Example 3: Tuple unpacking
point = (10, 20, 30)
x, y, z = point
print(f"x = {x}, y = {y}, z = {z}")

# Swapping values using tuples
a, b = 5, 10
print(f"Before swap: a = {a}, b = {b}")
a, b = b, a
print(f"After swap: a = {a}, b = {b}")
````

````python
# Example 4: Tuple methods
numbers = (1, 2, 3, 2, 4, 2, 5)

print(f"Count of 2: {numbers.count(2)}")
print(f"Index of 4: {numbers.index(4)}")
print(f"Length: {len(numbers)}")
````

## ðŸ’¡ Examples - Sets

````python
# Example 5: Creating sets
empty_set = set()
numbers = {1, 2, 3, 4, 5}
fruits = {"apple", "banana", "cherry"}

# Sets remove duplicates automatically
duplicates = {1, 2, 2, 3, 3, 3, 4}
print("With duplicates removed:", duplicates)

# Creating set from list
my_list = [1, 2, 2, 3, 4, 4, 5]
my_set = set(my_list)
print("From list:", my_set)
````

````python
# Example 6: Adding and removing elements
fruits = {"apple", "banana"}
print("Original:", fruits)

fruits.add("cherry")
print("After add:", fruits)

fruits.remove("banana")  # Raises error if not found
print("After remove:", fruits)

fruits.discard("grape")  # No error if not found
print("After discard (no error):", fruits)
````

````python
# Example 7: Set operations
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# Union: all elements from both sets
print("Union:", set_a | set_b)
print("Union (method):", set_a.union(set_b))

# Intersection: common elements
print("Intersection:", set_a & set_b)
print("Intersection (method):", set_a.intersection(set_b))

# Difference: elements in A but not in B
print("Difference A-B:", set_a - set_b)
print("Difference (method):", set_a.difference(set_b))
````

````python
# Example 8: Set membership testing
fruits = {"apple", "banana", "cherry"}

print("Is 'apple' in fruits?", "apple" in fruits)
print("Is 'grape' in fruits?", "grape" in fruits)
print("Is 'grape' not in fruits?", "grape" not in fruits)
````

````python
# Challenge 3 Solution
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
highest = students[0]
for student in students:
    if student[1] > highest[1]:
        highest = student
print(f"Highest scorer: {highest[0]} with {highest[1]} points")
````
